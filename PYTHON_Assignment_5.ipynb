```python
# Program to calculate string length using len()
my_string = input("Enter a string: ")
length = len(my_string)
print(f"The length of the string is: {length}")
```

```python
def get_string_ends(s):
    # Check if string length is less than 2
    if len(s) < 2:
        return ""
    # Combine first two and last two characters
    return s[:2] + s[-2:]

# Get input from the user
user_input = input("Enter a string: ")
result = get_string_ends(user_input)

print(f"Resulting string: {result}")
```

```python
# Program to join two strings
string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

# Use the + operator to join them
# Adding a space " " in between makes it readable
result = string1 + " " + string2

# Display the combined string
print("Concatenated string:", result)
```
