```python
my_list= [10,20,30,40]
print(20 in my_list)
print(60 in my_list)
print(60 not in my_list)
print(10 not in my_list)
```

```python
my_list = [10, 20, 30, 40, 50]

# Indexing
print(my_list[0])   # First element
print(my_list[2])   # Third element
print(my_list[-1])  # Last element (negative indexing)
print(my_list[-2])  # Second to last

# Slicing
print(my_list[1:4]) # Elements from index 1 to 3
print(my_list[:3])  # First three elements
print(my_list[2:])  # From index 2 to the end
```

```python
my_list = [10, 20, 30, 40]

# Update
my_list[1] = 25
print("After update:", my_list)

# Append
my_list.append(50)
print("After append:", my_list)

# Insert
my_list.insert(2, 15)
print("After insert:", my_list)

# Delete elements
my_list.remove(30)   # Removes specific value
print("After remove:", my_list)

my_list.pop()        # Removes last item
print("After pop:", my_list)

del my_list[1]       # Removes item at index 1
print("After del:", my_list)
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Concatenation
result = list1 + list2
print("Concatenation:", result)

# Repetition
print("Repetition:", list1 * 2)

# Built-in functions
print("Length:", len(result))
print("Maximum:", max(result))
print("Minimum:", min(result))
```
