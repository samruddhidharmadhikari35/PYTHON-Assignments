```python
# Using a while loop
i = 2
while i <= 10:
    print(i)
    i += 2
```

```python
odd_sum = 0
num = 1

while num <= 10:
    if num % 2 != 0:
        odd_sum += num
    num += 1

print("The sum is:", odd_sum)
```

```python
# Initialize the first two terms of the series
a, b = 0, 1

print("Fibonacci series between 0 and 50:")

# Loop until the current number 'a' exceeds 50
while a <= 50:
    print(a, end=" ")
    # Update 'a' to the current 'b', and 'b' to the sum of 'a' and 'b'
    a, b = b, a + b
```

```python
def remove_odd_index_loop(string):
    result = ""
    for i in range(len(string)):
        if i % 2 == 0:
            result += string[i]
    return result

# Example usage
input_str = "python"
print(remove_odd_index_loop(input_str)) # Output: "pto"
```
