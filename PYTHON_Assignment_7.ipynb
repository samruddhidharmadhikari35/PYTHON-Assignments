```python
# Create tuples
t1 = (1, 2, 3)
t2 = (4, 5, 6)

# Length
print("Length:", len(t1))

# Concatenation
t3 = t1 + t2
print("Concatenation:", t3)

# Repetition
t4 = t1 * 2
print("Repetition:", t4)

# Membership
print("Is 3 in t1?", 3 in t1)
print("Is 10 in t1?", 10 in t1)
```

```python
t = (10, 20, 30, 40, 50)

# Indexing and Negative Indexing
print("First element:", t[0])
print("Last element:", t[-1])

# Slicing
print("Slice (1 to 3):", t[1:4])

# Iteration
print("Iterating through tuple:")
for item in t:
    print(item)
```

```python
t = [1, 2, 3]

# Demonstrating Immutability
try:
    t[0] = 100  # This will raise an error
except TypeError as e:
    print("Error:", e)

# Demonstrating Deletion
del t
try:
    print(t) # This will raise an error because 't' no longer exists
except NameError as e:
    print("Confirmation:", e)
```

```python
seq = [10, 50, 20, 40] # A list sequence

# Convert sequence to tuple
t = tuple(seq)

print("Tuple:", t)
print("Length:", len(t))
print("Max value:", max(t))
print("Min value:", min(t))
```
