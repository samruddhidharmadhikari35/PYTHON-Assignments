```python
d = {'a': 3, 'b': 1, 'c': 2}

# Ascending
asc = dict(sorted(d.items(), key=lambda x: x[1]))
# Descending
desc = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))

print("Ascending:", asc)
print("Descending:", desc)
```

```python
d = {'a': 1, 'b': 2}
key = 'a'

if key in d:
    print(f"Key '{key}' exists")
else:
    print(f"Key '{key}' does not exist")
```

```python
d1 = {'a': 1}
d2 = {'b': 2}

# Using update() or dictionary unpacking
merged = {**d1, **d2}
print("Merged:", merged)
```
